```python
from bert_serving.client import BertClient
from sklearn.metrics.pairwise import cosine_similarity
bc = BertClient()
```

```python
import numpy as np
```

```python
#存储每个文件中 每个大分类的字符串
str_entcategory_text=dict()
for fileName in pred_docs.keys():
    str_entcategory_text[fileName]=[]
    for i in range(len(pred_docs[fileName].ents.ents)):
        string = pred_docs[fileName].ents.ents[i].category+" "+pred_docs[fileName].ents.ents[i].text
        str_entcategory_text[fileName].append(string)
str_entcategory_text        
```

```python
fw = open('str_entcategory_text.txt','wb')
pickle.dump(str_entcategory_text, fw)
fw.close()
```

```python
pre_bert = dict()
a=[]
for fileName in str_entcategory_text.keys():
    for i,j in enumerate(str_entcategory_text[fileName]):
        if i == 0:
            a = bc.encode([j]) 
        else:
            b = bc.encode([j]) 
            a = np.concatenate((a,b))
    pre_bert[fileName]=a
```

```python
fw = open('pre_bert.txt','wb')
pickle.dump(pre_bert, fw)
fw.close()
```

```python
label_name_doct=dict()
for filename in pre_bert.keys():
    
    final_list=[]
    for i in pre_bert[filename]:
        index_ = np.argmax(cosine_similarity([i], label_bert))
        final_list.append(detaillabel[index_])
    label_name_doct[filename] = final_list
    print(filename)
```
