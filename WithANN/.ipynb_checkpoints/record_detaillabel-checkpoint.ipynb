```python
#load file
def txt_strtonum_feed(filename):
    print(filename)
    data = []
    flag=True
    with open(filename, 'r', encoding='UTF-8') as f:#with auto call close()
        line = f.readline()
        
        while line:
            read_data = line.split('\n')
#             if flag:
#                 flag=False
#                 line = f.readline()
#                 continue
                
            read_data = read_data[0].split('\t')
            #"brat/12585968.ann"末尾存在不符合规则的内容
            if len(read_data) < 3:
                line = f.readline()
                continue
            
            first = read_data[0]
            second = read_data[1].split()[0]
            third=(read_data[1].split()[1],read_data[1].split()[2])
            forth = read_data[2]
            data.append((first,second,third,forth))
            line = f.readline()
        return data
```

```python
import pickle
fr2 = open('dataFilepred_docs.txt','rb')
data2 = pickle.load(fr2)
data2
```

```python
data2.keys()
```

```python
doct_detail_label = dict()
for filename in data2.keys():
    filename_ = "brat/"+filename+".ann"
    load_test_label = txt_strtonum_feed(filename_)
    doct_detail_label[filename]=dict()
    for i in load_test_label:
        doct_detail_label[filename][i[0]]=i[1]
```

```python
fw = open('doct_detail_label.txt','wb')
# Pickle the list using the highest protocol available.
#pickle.dump(dataList, fw, -1)
# Pickle dictionary using protocol 0.
pickle.dump(doct_detail_label, fw)
fw.close()
```
